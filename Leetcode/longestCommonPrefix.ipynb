```python
"""lets put in a class and test in leetcode"""

class Solution:
    def longestCommonPrefix(self, strs:List[str]) -> str: 
        strs = sorted(strs)
        first = strs[0]
        last = strs[-1]

        output_str = ""
        for words in range(len(first)):
            if first[words] != last[words]:
                return output_str
            else :
                output_str += first[words]
        return output_str
```
